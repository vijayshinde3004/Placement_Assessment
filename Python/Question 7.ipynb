Get all the Earth meteorites that fell before the year 2000
Get all the earth meteorites co-ordinates who fell before the year 1970
Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('nasa_data.csv') 
```

```python
data.shape
```

```python

data['year'] = pd.to_datetime(data['year'], errors='coerce')


earth_meteorites_before_2000 = data[data['reclat'].notnull() & (data['year'].dt.year < 2000)]


plt.figure(figsize=(10, 6))
sns.scatterplot(x='reclong', y='reclat', data=earth_meteorites_before_2000)
plt.title('Earth Meteorites Before 2000')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python

earth_meteorites_before_1970 = data[data['reclat'].notnull() & (data['year'].dt.year < 1970)]


plt.figure(figsize=(10, 6))
sns.scatterplot(x='reclong', y='reclat', data=earth_meteorites_before_1970)
plt.title('Earth Meteorites Before 1970')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python

earth_meteorites_mass_gt_10000 = data[data['mass'] > 10000]


plt.figure(figsize=(10, 6))
sns.histplot(data=earth_meteorites_mass_gt_10000, x='mass', bins=20)
plt.title('Earth Meteorites with Mass > 10000 kg')
plt.xlabel('Mass')
plt.ylabel('Count')
plt.show()
```
